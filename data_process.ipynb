```python
import numpy as np
import pandas as pd
```

```python
file_path_2020 = r"E:\Dataset\Load_Europe\monthly_hourly_load_values_2020.csv"
file_path_2021 = r"E:\Dataset\Load_Europe\monthly_hourly_load_values_2021.csv"
file_path_2022 = r"E:\Dataset\Load_Europe\monthly_hourly_load_values_2022.csv"
file_path_2023 = r"E:\Dataset\Load_Europe\monthly_hourly_load_values_2023.csv"
file_path_2024 = r"E:\Dataset\Load_Europe\monthly_hourly_load_values_2024.csv"
# Load each file with the correct separator
df_2020 = pd.read_csv(file_path_2020, sep='\t')
df_2021 = pd.read_csv(file_path_2021,sep=';')
df_2022 = pd.read_csv(file_path_2022,sep=';')
df_2023 = pd.read_csv(file_path_2023,sep='\t')  # Assuming this is TSV format
df_2024 = pd.read_csv(file_path_2024)  # Assuming this is TSV format

# Ensure the 'DateUTC' column is in datetime format
df_2020['DateUTC'] = pd.to_datetime(df_2020['DateUTC'], format='%d-%m-%Y %H:%M')
df_2021['DateUTC'] = pd.to_datetime(df_2021['DateUTC'], format='%d/%m/%Y %H:%M')
df_2022['DateUTC'] = pd.to_datetime(df_2022['DateUTC'], format='%d/%m/%Y %H:%M')
df_2023['DateUTC'] = pd.to_datetime(df_2023['DateUTC'], format='%d-%m-%Y %H:%M')
df_2024['DateUTC'] = pd.to_datetime(df_2024['DateUTC'], format='%d/%m/%Y %H:%M')

# Select only relevant columns for concatenation
df_2020_selected = df_2020[['DateUTC', 'CountryCode', 'Value']]
df_2021_selected = df_2021[['DateUTC', 'CountryCode', 'Value']]
df_2022_selected = df_2022[['DateUTC', 'CountryCode', 'Value']]
df_2023_selected = df_2023[['DateUTC', 'CountryCode', 'Value']]
df_2024_selected = df_2024[['DateUTC', 'CountryCode', 'Value']]

# Concatenate the dataframes from 2020 to 2024
combined_df_all_years = pd.concat([df_2020_selected, df_2021_selected, df_2022_selected, df_2023_selected, df_2024_selected], ignore_index=True)

# Save the combined dataset to a CSV file
combined_file_path = 'combined_load_data_2020_2024.csv'
combined_df_all_years.to_csv(combined_file_path, index=False)

print(f"Combined data saved to {combined_file_path}")
```
